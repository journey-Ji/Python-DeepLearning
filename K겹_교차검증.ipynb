하나의 dataset을 테스트셋, 학습셋으로 나누면 학습의 양도 떨어지고, 테스트의 정확도도 떨어집니다.

따라서 이를 해결하기 위한 것이 K겹 교차검증입니다.

데이터셋을 여러개로 나누어서 하나씩 테스트셋으로 사용하고 나머지를 전부 합해서 학습셋으로 이용합니다.

sklearn의 StratiifiedKFold()함수를 이용합니다.

```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import numpy
import pandas as pd
import tensorflow as tf

#set seed
seed = 0 
numpy.random.seed(seed)
tf.random.set_seed(seed)
df = pd.read_csv('sonar.csv',header=None)

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)


#10개로 쪼갬
n_fold =10
skf =StratifiedKFold(n_splits=n_fold, shuffle = True, random_state =seed) # 10개로 쪼개어서 섞어주는 함수설정

#빈 배열
accuracy =[]

# 모델 설정, 컴파ㅏ일, 실행 반복
for train,test in skf.split(X,Y): #X,Y를 테스트와 학습셋으로 쪼갬
  model = Sequential()
  model.add(Dense(24,input_dim=60, activation='relu'))
  model.add(Dense(10,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='mean_squared_error',
                optimizer='adam',
                metrics=['accuracy'])
  model.fit(X[train],Y[train],epochs=100,batch_size=5) #학습셋으로 학습
  k_accuracy="%.4f"%(model.evaluate(X[test],Y[test])[1]) #테스트셋으로 테스트
  accuracy.append(k_accuracy) #배열에 결과 저장


#print
print("\n %.f fold accuracy : "%n_fold,accuracy)
```
